```python
import numpy as np
import matplotlib.pyplot as plt


vals = [1, 0.1, 0.01, 0.001, 0.0001]

for lr in vals:
    l = np.fromfile(f'LR_tests/TE_{lr}.dat')
    plt.plot(l, label=f'{lr}')

plt.title("Training error for batches of 500")
plt.legend()
plt.yscale('log')
plt.xlim(0,80)
plt.show()




for lr in vals:
    l = np.fromfile(f'LR_tests/LOSS_{lr}.dat')
    plt.plot(l, label=f'{lr}')

plt.title("Loss for batches of 500")
plt.legend()
# plt.yscale('log')
plt.show()


```

# Full batch avg

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values for LR
vals = [15, 40, 110, 510]

# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
plt.suptitle("Training error for full-batch averages on different batch sizes")

for i, lr in enumerate(vals):
    ax = axes[i // 2, i % 2]  # Determine the subplot index
    l = np.fromfile(f'LR_tests/TE_BATCH_AVG_{lr}_0.0001.dat')
    ax.plot(l, label=f'{lr}')
    ax.set_title(f'Training error for batch size = {lr}')
    ax.set_xlabel('x-axis (steps)')
    ax.set_ylabel('Training error')

# Ensure that all subplots have labels and adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()


```
